```python
import control as ctl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# sistema de 1 ordem
R=10e3;C=20e-6; 
#cria a função de transferência
s = ctl.TransferFunction.s
G_s  = (1)/(R*C*s + 1)
print(' G_s = ',G_s)
#diagrama de bode
Ws=np.logspace(-3,3,1000);
plt.figure(1)
ctl.bode_plot(G_s, Ws,dB=True)
plt.xlabel('w (rad/s)')
```

```python
# sistema de 2 ordem x 1 ordem
plt.figure(3)
quais_zeta = [0.1,0.3,1,1.5];
wn=1;
Ws=np.logspace(-2,2,1000);
for i in range(0,len(quais_zeta)):
  G_s  = (s**2+2*quais_zeta[i]*wn*s + wn**2)
  #print(' G_s = ',G_s)
  bp=ctl.bode_plot(G_s, Ws,dB=True,wrap_phase=True)
plt.xlabel('w/Wn')
plt.legend(['$\zeta = 0.1$','$\zeta = 0.3$','$\zeta = 1$','$\zeta = 1.5$'])
```

```python
# sistema de 2 ordem
zeta = 0.4;
wn=20;
a=5;
Ws=np.logspace(-1,3,1000)
plt.figure(2)
G2_s=(s+a);
G3_s  = 1/(s**2+2*zeta*wn*s + wn**2)
G1_s = G2_s*G3_s;
print(' G1_s = ',G1_s)
bp=ctl.bode_plot(G2_s, Ws,dB=True,wrap_phase=True)
bp=ctl.bode_plot(G3_s, Ws,dB=True,wrap_phase=True)
bp=ctl.bode_plot(G1_s, Ws,dB=True,wrap_phase=True)
plt.legend(['$G_1(s)$','$G_2(s)$','$G(s)$'])
 
```
